```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#RMSE Function
def rmse_calculation(actual, predicted):
    actual = np.array(actual)
    predicted = np.array(predicted)
    squared_diff = (actual - predicted) ** 2
    mean_squared_diff = np.mean(squared_diff)
    rmse = np.sqrt(mean_squared_diff)
    return rmse

#MAE Function
def mae_calculation(actual, predicted):
    actual = np.array(actual)
    predicted = np.array(predicted)
    absolute_diff = np.abs(actual - predicted)
    mae = np.mean(absolute_diff)
    return mae

#Accuracy Function
def accuracy_calculation(actual, predicted):
    actual = np.array(actual)
    predicted = np.array(predicted)
    correct_predictions = np.sum(actual == predicted)
    accuracy = correct_predictions / len(actual) * 100
    return accuracy

#Housing Data
housing_data = pd.read_csv('housing_data.csv')

#Housing Data Calculations
rmse_value = rmse_calculation(housing_data['sale_price'], housing_data['sale_price_pred'])
mae_value = mae_calculation(housing_data['sale_price'], housing_data['sale_price_pred'])

print(f"Housing Data RSME: {rmse_value}")
print(f"Housing Data MAE: {mae_value}")

#Mushroom Data
mushroom_data = pd.read_csv('mushroom_data.csv')

#Mushroom Data Calculation
accuracy_value = accuracy_calculation(mushroom_data['actual'], mushroom_data['predicted'])

print(f"Mushroom Data Accuracy: {accuracy_value}%")

#Error Function
def error_function(p):
    return 0.005*p**6 - 0.27*p**5 + 5.998*p**4 - 69.919*p**3 + 449.17*p**2 - 1499.7*p + 2028

p_values = np.linspace(0, 10, 500)
error_values = error_function(p_values)

#Plot Error Function
plt.figure(figsize=(10, 6))
plt.plot(p_values, error_values, label='Error Function')
plt.title('Error Function Plot')
plt.xlabel('p')
plt.ylabel('Error')
plt.grid(True)
plt.legend()
plt.show()

min_error = np.min(error_values)
min_p = p_values[np.argmin(error_values)]

print(f"Optimal p-value: {min_p}")
print(f"Minimum Error: {min_error}")
```

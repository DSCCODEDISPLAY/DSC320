```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x):
    return np.sqrt(4 - x ** 2)

#Step A: Sketch a graph of f and the rectangles that make up the right Riemann sum R4 on the interval [0, 2].
a = 0
b = 2
n = 4
delta_x = (b - a) / n

x = np.linspace(a, b, 100)
y = f(x)

rect_x = np.linspace(a + delta_x, b, n)
rect_y = f(rect_x)

plt.plot(x, y, label=r'$f(x) = \sqrt{4 - x^2}$')

for i in range(n):
    plt.bar(rect_x[i], rect_y[i], width=delta_x, align='edge', alpha=0.2, color='orange', edgecolor='black')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Right Riemann Sum with 4 Rectangles')
plt.legend()
plt.grid(True)
plt.show()

#Step B: Use the function you created in the previous problem to calculate R4.
def right_riemann_sum(f, a, b, n):
    delta_x = (b - a) / n
    total_sum = 0

    for k in range(1, n + 1):
        x_k = a + k * delta_x
        total_sum += f(x_k)

    return total_sum * delta_x

R4 = right_riemann_sum(f, a, b, n)
print(f"R_4 = {R4}")

#Step C: Calculate appropriate right Riemann sums to find the area under the curve y = f(x) on the interval [0, 2].
area, _ = quad(f, a, b)
print(f"Exact area under the curve = {area}")
print("The area under the curve represents one-quarter of the area of a circle with radius 2, and the exact area is Ï€.")
```

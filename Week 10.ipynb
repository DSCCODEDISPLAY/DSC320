```python
#Step 1: Summarizing Data
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('qb_stats.csv')

#Step 1(a): Find the mean of each numerical column of data.
mean_values = data.mean(numeric_only=True)
print("Step 1(a):")
print("Mean of Each Numerical Column")
print(mean_values)

#Step 1(b): Find the standard deviation of each numerical column of data.
std_dev_values = data.std(numeric_only=True)
print("\nStep 1(b):")
print("Standard Deviation of Each Numerical Column")
print(std_dev_values)

#Step 1(c): Create a histogram of the number of yards; label it appropriately.
print("\nStep 1(c):")
plt.figure(figsize=(10, 6))
plt.hist(data['yds'], bins=5, edgecolor='black')
plt.title('Histogram of Number of Yards')
plt.xlabel('Number of Yards')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#Step 1(d): Create a boxplot of the number of touchdowns. Identify any outliers.
print("\nStep 1(d):")
plt.figure(figsize=(10, 6))
plt.boxplot(data['td'], vert=False, patch_artist=True)
plt.title('Boxplot of Number of Touchdowns')
plt.xlabel('Number of Touchdowns')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

touchdowns = data['td']
q1 = touchdowns.quantile(0.25)
q3 = touchdowns.quantile(0.75)
iqr = q3 - q1
outliers = touchdowns[(touchdowns < (q1 - 1.5 * iqr)) | (touchdowns > (q3 + 1.5 * iqr))]

print("\nOutliers in the number of touchdowns:")
print(outliers)
```

```python
#Step 2: Calculating Probabilities from Data
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('survey_data.csv')

#Step 2(a): Based off of this data, what is the probability a college student has brown hair?
p_brown_hair = (data['hair_color'] == 'brown').mean()
print(f"Step 2(a): Probability a student has brown hair is {p_brown_hair:.2%}.")

#Step 2(b): Based off of this data, what is the probability a college student has blue eyes?
p_blue_eyes = (data['eye_color'] == 'blue').mean()
print(f"Step 2(b): Probability a student has blue eyes is {p_blue_eyes:.2%}.")

#Step 2(c): Based off of this data, what is the probability a college student has blue eyes given that they have brown hair?
brown_hair_data = data[data['hair_color'] == 'brown']
p_blue_given_brown = (brown_hair_data['eye_color'] == 'blue').mean()
print(f"Step 2(c): Probability a student has blue eyes given they have brown hair is {p_blue_given_brown:.2%}.")

#Step 2(d): Based off of this data, what is the probability a college student has brown hair given that they have blue eyes?
blue_eyes_data = data[data['eye_color'] == 'blue']
p_brown_given_blue = (blue_eyes_data['hair_color'] == 'brown').mean()
print(f"Step 2(d): Probability a student has brown hair given they have blue eyes is {p_brown_given_blue:.2%}.")

#Step 2(e): Do your results above indicate that college students having brown hair and blue eyes are independent of one another? Explain.
p_brown_and_blue = ((data['hair_color'] == 'brown') & (data['eye_color'] == 'blue')).mean()
independence_check = p_brown_hair * p_blue_eyes

print("\nStep 2(e):")
print(f"P(Brown Hair and Blue Eyes): {p_brown_and_blue:.2%}")
print(f"P(Brown Hair) * P(Blue Eyes): {independence_check:.2%}")
if abs(p_brown_and_blue - independence_check) < 1e-5:
    print("Brown Hair and Blue Eyes are independent.")
else:
    print("Brown Hair and Blue Eyes are not independent.")

#Step (f): Create a bar graph of the hair color and eye color of this group of students. Label the graphs appropriately.
print("\nStep 2(f):")
plt.figure(figsize=(8, 5))
data['hair_color'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Distribution of Hair Color')
plt.xlabel('Hair Color')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
data['eye_color'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Distribution of Eye Color')
plt.xlabel('Eye Color')
plt.ylabel('Frequency')
plt.show()
```

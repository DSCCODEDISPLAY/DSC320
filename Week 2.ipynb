```python
#Step 1: Make a scatterplot of the highway miles per gallon (y-axis) versus the weight (x-axis).
import pandas as pd
import matplotlib.pyplot as plt

car_data = pd.read_csv('car_data.csv')

x = car_data['weight']
y = car_data['hwy_mpg']

plt.scatter(x, y, alpha=0.5)
plt.xlabel('Weight of the Car (lbs)')
plt.ylabel('Highway Miles per Gallon (MPG)')
plt.title('Scatterplot of Highway MPG vs. Car Weight')
plt.show()
```

```python
#Step 6: Use Python to find the best-fit line. The Scikit-learn package is a good choice to use for this.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

car_data = pd.read_csv('car_data.csv')

x = car_data['weight'].values.reshape(-1, 1)
y = car_data['hwy_mpg']

model = LinearRegression()
model.fit(x, y)

slope = model.coef_[0]
y_intercept = model.intercept_

y_prediction = model.predict(x)

print(f"Best-Fit Slope: {slope}")
print(f"Best-Fit Y-Intercept: {y_intercept}")

plt.scatter(x, y, alpha=0.5)
plt.plot(x, y_prediction, color='red', label=f'Best Fit: y = {slope:.4f}x + {y_intercept:.4f}')
plt.xlabel('Weight of the Car (lbs)')
plt.ylabel('Highway Miles per Gallon (MPG)')
plt.title('Scatterplot of Highway MPG vs. Car Weight')
plt.legend()
plt.show()
```

```python
#Step 7: Find the root mean squared error (RMSE) of the prediction line you found in problem (4) and the actual best-fit line found in problem (5). How do these compare? 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

car_data = pd.read_csv('car_data.csv')

x = car_data['weight']
y = car_data['hwy_mpg']

slope_manual = -0.007
y_intercept_manual = 50
manual_predictions = slope_manual * x + y_intercept_manual

X = x.values.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)

slope_best_fit = model.coef_[0]
intercept_best_fit = model.intercept_
best_fit_predictions = model.predict(X)

manual_rmse = np.sqrt(mean_squared_error(y, manual_predictions))
best_fit_rmse = np.sqrt(mean_squared_error(y, best_fit_predictions))

print(f"Manual Fit Slope: {slope_manual}")
print(f"Manual Fit Intercept: {y_intercept_manual}")
print(f"Manual Fit RMSE: {manual_rmse}")
print()
print(f"Best Fit Slope: {slope_best_fit}")
print(f"Best Fit Intercept: {intercept_best_fit}")
print(f"Best Fit RMSE: {best_fit_rmse}")


plt.scatter(x, y, alpha=0.5, label="Data Points")
x_line_manual = np.linspace(min(x), max(x), 100)
y_line_manual = slope_manual * x_line_manual + y_intercept_manual
plt.plot(x_line_manual, y_line_manual, color='red', label=f'Manual Line: y = {slope_manual}x + {y_intercept_manual}')

x_line_best_fit = np.linspace(min(x), max(x), 100)
y_line_best_fit = slope_best_fit * x_line_best_fit + intercept_best_fit
plt.plot(x_line_best_fit, y_line_best_fit, color='green', label=f'Best Fit Line: y = {slope_best_fit:.4f}x + {intercept_best_fit:.4f}')

plt.xlabel('Weight of the Car (lbs)')
plt.ylabel('Highway Miles per Gallon (MPG)')
plt.title('Scatterplot of Highway MPG vs. Car Weight')
plt.legend()
plt.show()
```
